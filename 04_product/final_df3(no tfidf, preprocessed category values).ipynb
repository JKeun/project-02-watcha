```python
pd.set_option('display.max_columns', None)
df = pd.read_csv('../resource/preprocess_df2.csv')
```

### category value 하나의 str샘플로 묶는 과정

```python
cat_X = df.ix[:,'director':'nation']
cat_X
```

```python
a = cat_X.director
b = cat_X.actors.str.replace(",", " ").str.replace(".", "").str.replace("-", "")
c = cat_X.film_rate.str.replace(" ","")
d = cat_X.genre.str.replace("/", " ")
e = cat_X.nation.str.replace(",", "")

text_X = a+" "+b+" "+c+" "+d+" "+e
text_X
```

```python
real_X = df.drop(['director', 'actors', 'film_rate', 'genre', 'nation'], axis=1)
```

```python
real_X['text'] = text_X
```

```python
real_X.to_csv("../resource/final_df3.csv")
```

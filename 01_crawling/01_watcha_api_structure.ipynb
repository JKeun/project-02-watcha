## 왓챠 json구조 분석
* 뽑아야 할 feature들

```python
import requests
import json
import pandas as pd
```

```python
response = requests.get("https://watcha.net/v2/users/jXaIHl0ZtdYZ/movies.json?filter%5Bsorting%5D=time&page=1")
```

```python
watcha_dict = json.loads(response.text)
watcha_dict
```

```python
watcha_dict.keys()
```

### 'cards'안에 원하는 정보존재
* 24개의 영화정보 존재

```python
watcha_list = watcha_dict.get('cards')

print(len(watcha_list))

watcha_list
```

```python
# 첫번째 영화(watcha_list[0]) 안에 있는
# 'items' key 안에 있는
# 'item' key 로 들어감
watcha_list[0].get('items')[0].get('item')
```

```python
watcha_list[23].get('items')[0].get('item').get('owner_action').get('rating')
```
